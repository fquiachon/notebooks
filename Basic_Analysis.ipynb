```python
import pandas as pd
oo = pd.read_csv("D:/my_workspace/pythons/data/olympics.csv", encoding='ISO-8859-1')
```

Basic Analysis
- Exercise 1 In which events did Jesse Owens win a medal?

```python
oo[oo.Athlete.str.contains('Owens') == True]
```

- Exercise 2 :  Which country has won the most men's gold medals in singles badminton over the years? Sort the results alphabetically by the players' names.

```python
mbg = oo[(oo.Sport == 'Badminton') & (oo.Medal == 'Gold') & (oo.Gender == 'Men')]
mbg.Country.value_counts()
```

```python
result = mbg.sort_values(by=['Athlete'])
result[['Athlete', 'Gender', 'Country']]
```

- Exercise 3 :  Which three countries have won the most medals in recent years, that's from 1984 to 2008?

```python
rec = oo[oo.Year >= 1984]
rec.head()
```

```python
rec.Country.value_counts().head(3)
```

- Exercise 4:  Display the male gold medal winners of the 100-meters track and field sprint event over the years. List the results starting with the most recent first and show the Olympic city, edition, athlete and the country they represent.

```python
mgs = oo[ (oo.Gender == 'Men') & (oo.Medal == 'Gold') & (oo.Event == '100m')]
sorted_mgs= mgs.sort_values(by=['Year'], ascending=False)
sorted_mgs[['City', 'Year', 'Athlete', 'Country']]
```
